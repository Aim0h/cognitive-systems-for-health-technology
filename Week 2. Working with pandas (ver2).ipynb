```python
# Import libraries
# %pylab inline
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
```

```python
# location of the dataset
filename = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.switzerland.data'

# column names for the data
colnames = ['age','sex','cp','tresbps','chol',
            'fbs','restecg','thalach','exang',
            'oldpeak','slope','ca','thal','num']

# read data to data frame
df = pd.read_csv(filename, 
                 names = colnames, 
                 na_values = '?')

df.head()
```

```python
df.tail()
```

```python
df.mean() # df.std, df.min, df.max, df.count, df.median, df.mode
```

```python
df.describe()
```

```python
df['age'].plot()
plt.xlabel('Index')
plt.ylabel('Age (Years)')
plt.grid()
plt.show()
```

```python
df['age'].hist(bins = np.arange(30, 100, 5))
plt.xlabel('Age (Years)')
plt.ylabel('Count')
plt.show()
```

```python
df.boxplot(column='age', by='sex')
plt.show()
```
