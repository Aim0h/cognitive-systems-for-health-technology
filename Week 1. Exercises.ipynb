```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create an array from 0 to 1000
x = np.arange(1001)
x
```

```python
# Calculate log10(x+1)
y = np.log10(x + 1)
y
```

```python
# Show the graph of y = log10(x+1)
plt.plot(x, y)
plt.title('Graph of y = log(x+1)')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.grid()
plt.show()
```
